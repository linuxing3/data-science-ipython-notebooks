```python
%matplotlib inline
from scipy import stats
import matplotlib.pylab as plt
import numpy as np
import pandas as pd
```

```python
plt.rc('figure',figsize=(10,5))
x = np.linspace(0, 2, 10)
plt.plot(x, x, 'o-', label='linear')
plt.plot(x, x ** 2, 'x-', label='quadratic')
plt.legend(loc='best')
plt.title('Linear vs Quadratic progression')
plt.xlabel('Input')
plt.ylabel('Output');
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
distribution = stats.t.rvs(10,size=100)
ax.hist(distribution,bins=100,normed=True)
plt.legend(loc='best')
plt.title('Histogram of random variantes')
plt.xlabel('Input')
plt.ylabel('Output');
plt.show()
```

```python
a, b = 2.31, 0.627
mean, var, skew, kurt = stats.beta.stats(a, b, moments='mvsk')
```

```python
print(stats.beta.ppf(0.01, a, b))
print(stats.beta.ppf(0.99, a, b))
```

```python
x = np.linspace(stats.beta.ppf(0.01, a, b),
                stats.beta.ppf(0.99, a, b), 100)
x
```

```python
plt.plot(x, stats.beta.pdf(x, a, b), 'go-', label='line 1', linewidth=2)
```
